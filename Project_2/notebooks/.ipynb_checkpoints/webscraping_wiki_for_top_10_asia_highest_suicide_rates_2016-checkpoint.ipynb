```python
from __future__ import print_function, division
import requests
```

```python
requests.__path__
```

```python
url = "https://en.wikipedia.org/wiki/List_of_countries_by_suicide_rate"
response = requests.get(url)
response.status_code  # status code = 200 = ok
```

## Look at the text

```python
response.text
```

## Soupify the text

```python
page = response.text
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(page, 'lxml')
```

```python
print(soup)
```

## Prettify the Soup

```python
print(soup.prettify())
```

```python
type(soup)
```

## BeautifulSoup - Find & Find_All  
  
### `soup.find()`

```python
# Finds the first instance of the tag <table>
soup.find('table')
```

```python
# Finds the next instance of the tag <table>
soup.find('table').findNextSibling()
```

```python
# Upon inspection, the table of importance (the one we want to extract from) lies in the third element of the list
soup.find_all('table')
```

```python
table_containing_suicide_rate_data = soup.find_all('table')[2]
```

```python
rows_with_suicide_data = table_containing_suicide_rate_data.find_all('tr')[2:]
rows_with_suicide_data
```

```python
# First item in the iterable list, rows_with_suicide_data
rows_with_suicide_data[0]
```

## Retrieving Continent

```python
country_data_with_td_tag = rows_with_suicide_data[0].find_all('td')[2]
country_data_with_td_tag
```

```python
country = country_data_with_td_tag.text[:-1]
country
```

## Retrieving Country

```python
rows_with_suicide_data[0].find_all('a')[0]['href']
```

```python
rows_with_suicide_data[0].find_all('a')[0]['href'].strip().split('/')[-1]
```

## Retrieving Both sexes num

```python
both_sexes_data_with_tag = rows_with_suicide_data[0].find_all('td')[3]
both_sexes_data_with_tag
```

```python
both_sexes = float(both_sexes_data_with_tag.text[:-1])
both_sexes
```

```python
# Scraped the name for every country in the Continent Column
both_sexes = []
countries = []
other_continents = []
asia_continent = []

for item in rows_with_suicide_data:
    continent = item.find_all('td')[2].text[:-1]
    if continent == "Asia":
        both_sexes.append(float(item.find_all('td')[3].text[:-1]))
        countries.append(item.find_all('a')[0]['href'].strip().split('/')[-1])
        asia_continent.append(continent)
    else:
        other_continents.append(continent)

full_list = list(zip(countries, both_sexes))
print(full_list, '\n')
print(asia_continent, '\n')
print(other_continents)
```
